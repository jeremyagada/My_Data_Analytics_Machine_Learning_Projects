```python
#import all the libraries and packages:
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
#We are going to import the 
df = pd.read_csv('movies.csv');

#Let's check if out dataset load correctly by getting the first 10 roles in the data set
df.head()
```

```python
#Now we are checking out Dataset for missing values
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print ('{} - {}%'.format (col, pct_missing))
```

```python
#We are going to be sorting the dataset by gross
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
df.head()
```

```python
#let's drop all the duplicate data
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
#We are going to check what elements are most correlated with gross revenue. We will be testing
#Budget high correlation
#Company High Correlation
```

```python
# We will be compare the budget vs gross using a scatter plot
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earning')
plt.ylabel('Budget for the Film')
plt.show
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numneric Features')
plt.xlabel('Movie Feature')
plt.ylabel('Movie Feature')
plt.show
```

```python
#We are going to be looking at non numeric values that are could be important to our analysis.
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
#Now we are going to be using the heatmap to check for the correlation between all the columns
df_numerized = df.corr(method='pearson')

sns.heatmap(df_numerized, annot=True)
plt.title('Correlation Matrix for Numneric Features')
plt.xlabel('Movie Feature')
plt.ylabel('Movie Feature')
plt.show
```

```python
#We are going to unstack this to see the columns with high correlation
correlation_matrix = df_numerized.corr()
corr_pairs = correlation_matrix.unstack()
corr_pairs.head(10)
```

```python
#let's sort the vlaues
sorted_pairs = corr_pairs.sort_values()

sorted_pairs.head()
```

```python
#let bring out the values with the highest correlation
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
